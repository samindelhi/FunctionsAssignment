1. What is the difference between a function and a method in Python?
 - Both *functions* and *methods* are are block of code mean to perform a specific task but they differe in the way there are called, the scope in which they are accessible and the object with which they are associated.

```
    def greet(name):

    print(greet("Shyam"))  # Output: Hello, Shyam!
```
  - Methods: A method, on the other hand, is a function that belongs to an object (usually an instance of a class). It is defined within a class and operates on the data contained within that object(the instance of that class).
```

g = Wisher
g.greet("Shyam")
```

  numbers = [1, 2, 3, 4, 5]  # List (Iterable)

6. Explain the concept of generators in Python and how they are defined.
  - Generators in Python, are a special type of iterator that allow for lazy evaluation—meaning they produce values one at a time on demand instead of storing them in memory all at once. This makes them useful especially while working with large datasets or streaming data efficiently.
    - Defining a Generator:
      - Generators are defined same way as a function using the def keyword but with the **yield** keyword instead of **return**. Unlike regular functions, they don’t execute completely; instead, they pause at yield, returning a value while retaining their state for the next call.
```
    def count_up_to(n):
        count = 1
        while count <= n:
            yield count  # Pauses here and returns count
            count += 1
    counter = count_up_to(5)    # Creating a generator object
    #Iterating over generator values
    for num in counter:
        print(num)  # Output: 1, 2, 3, 4, 5
```
      - How Generators Work
          - When yield is encountered, the function execution pauses.
          The next value is produced only when next() is called on the generator.
          Generators retain their state, avoiding the need for recomputation.
        Example using next():
```
        gen = count_up_to(3)
        print(next(gen))  # Output: 1
        print(next(gen))  # Output: 2
        print(next(gen))  # Output: 3
        print(next(gen)) # when next(gen) is called again after all values are yielded, a StopIteration error occurs.
```

9. Explain the purpose and usage of the `map()` function in Python.
  - The map() function in Python is used to apply a function to every item in an iterable (e.g., lists, tuples) without needing explicit loops. It's a functional programming tool that enhances readability, efficiency and reduces lines of code.
  - Purpose:
    1. Applies a function to all elements of an iterable.
    2. Avoids manual loops by automating transformation.
    3. Enhance performance, especially while executing large datasets.
```
    numbers = [1, 2, 3, 4, 5]
    squared_numbers = map(lambda x: x ** 2, numbers) #retuns a map object
    # Here, lambda x: x ** 2 applies the function to each number in numbers.
    print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]

        # using map() with multiple iterables
    a = [1, 2, 3]
    b = [4, 5, 6]
    sum_values = map(lambda x, y: x + y, a, b)
    print(list(sum_values))  # Output: [5, 7, 9]
```
  - Key Benefits of map()
    - Improves readability by eliminating explicit loops.
    - Enhances performance since transformations are done lazily.
    - Works well with lambda functions for concise expressions.

|Function|Description|Returns|

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    """Check if a number is prime."""
    if n < 2:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False  # Found a divisor, so not prime
    return True  # No divisors found, it's prime

# Checking prime numbers in the range 1 to 200
prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
````

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class fibIterator:
  def __init__(self,nterms):
    self.nterms = nterms
    self.a, self.b = 0,1

  def __iter__(self):
    return(self)    # Returns the iterator instance

  def __next__(self):
    if self.nterms <= 0:
      raise StopIteration #Stops iterating when the limit is reached.
    self.nterms -=1     #decrementing the number
    result = self.a
    self.a,self.b = self.b,self.a+self.b # Fibonacci step
    return result

fib = fibIterator(21)
print(list(fib))
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def genfunc(exp):
  """Generator to yield powers of 2 up to 2^exp."""
  for x in range(1, exp + 1): #Starts at 2^1 upto and Including 2^exp
    yield 2 ** x
#Usage
gen = genfunc(5)  #Generates 2^1 to 2^3
print(list(gen))  #Output:  [2, 4, 8, 16, 32]
````
